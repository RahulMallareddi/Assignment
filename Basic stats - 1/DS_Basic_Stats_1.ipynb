# Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

```python
#Importing all the required packages and reading the sales_data_with_discounts.csv for computation

import pandas as pd
import numpy as np
from scipy import stats as st
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
# convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
# Creating a list of numerical and non-numerical column separately
numerical_colums_heading = df.select_dtypes(include=np.number).columns.tolist()
non_numerical_colums_heading = df.select_dtypes(exclude=np.number).columns.tolist()
```

## Descriptive Analytics for Numerical Columns
●	Objective: To compute and analyze basic statistical measures for numerical columns in the dataset.

●	Steps:
	Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
	Identify numerical columns in the dataset.
	Calculate the mean, median, mode, and standard deviation for these columns.
	Provide a brief interpretation of these statistics.

```python
# Creating a function to give Mean, Median, Mode and Standard Deviation of numerical column
def descriptive_analysis(data_array):
    print("Mean: ",np.mean(data_array))
    print("Median: ",np.median(data_array))
    print("Mode: ",st.mode(data_array))
    print("Standard Deviation: ",np.std(data_array))
```

```python
# Looping through descriptive_analysis function to calculate the Mean, Median, Mode and Standard Deviation 
#for each numerical columns
for each in numerical_colums_heading:
    print("================This is the Descriptive Analytics for column :",each,"==========================")
    descriptive_analysis(df[each])
    print("================================================================================================")
```

#### Brief interpretation of these statistics.

**Mean**
The mean (also know as average), is obtained by dividing the sum of observed values by the number of observations, n.
*Mean provides a quick overview of all trends or typical values of the data.*

**Median**
The median is the middle value of a set of data containing an odd number of values, or the average of the two middle values of a set of data with an even number of values.
*The median is especially helpful when separating data into two equal sized bins.*

**Mode**
The mode of a set of data is the value which occurs most frequently.
*Helps to find the most trending data in a column.*

**Standard Deviation**
The standard deviation gives an idea of how close the entire set of data is to the average value.
*Data sets with a small standard deviation have tightly grouped, precise data. Data sets with large standard deviations have data spread out over a wide range of values.*

## Boxplots
Create boxplots for numerical variables to identify outliers and the interquartile range.

Discuss any findings, such as extreme values or unusual distributions.


```python
def box_plotting(column_heading):
    plt.figure(figsize=(20,5))
    plt.boxplot(df[column_heading],vert = False)
    plt.xlabel(column_heading)
```

```python
for each in numerical_colums_heading:
    box_plotting(each)
```

**Discuss any findings, such as extreme values or unusual distributions.**

It's clear from the above box plots that each numerical column have extreme value/outliers. An outlier is a single data point that goes far outside the average value of a group of statistics. Each and every dot in box plot represents the outliers.

Below function will find all outliers in numerical columns.

```python
#create a function to find outliers using IQR

def find_outliers(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers
```

```python
#Looping through find_outliers function to find outliers/extreme values in each numerical columns.

for each in numerical_colums_heading:
    print("================Finding the outliers for column :",each,"==========================")
    outliers = find_outliers(df[each])
    print("number of outliers: "+ str(len(outliers)))
    print("max outlier value: "+ str(outliers.max()))
    print("min outlier value: "+ str(outliers.min()))
    print(outliers)
    print("====================================================================================")
```

# Standardization of Numerical Variables
●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

●	Steps:
    
    Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma
	Show before and after comparisons of the data distributions.


```python
df1=df.copy()
df1
```

```python
#Creating a function to calcuate the z-scores for all the numerical column

def calculate_Z_Scores(column_heading):
    df1['z_'+column_heading]=(df1[column_heading]-np.mean(df1[column_heading]))/np.std(df1[column_heading])
```

```python
# Looping through calculate_Z_Scores function to calcuate the z-scores for all the numerical columns

for each in numerical_colums_heading:
    calculate_Z_Scores(each)
```

```python
df1
```

```python
# Creating a function to create a hist plot (KDE=True) with Original and corresponding Z-values for same numerical column.

def histogram_plotting_Before_And_After_Z_Score(column_heading):
    fig, ax = plt.subplots(1,2, figsize=(15,5))
    sns.histplot(df1[column_heading], ax=ax[0], bins=10,kde=True)
    sns.histplot(df1['z_'+column_heading], ax=ax[1], bins=10,kde=True)
```

```python
# Looping through histogram_plotting_Before_And_After_Z_Score function for each numerical columns.

for each in numerical_colums_heading:
    histogram_plotting_Before_And_After_Z_Score(each)
```

Z scores indicate the position of data values by measuring their distance from the mean. Z score gives an idea of where the data points fit into the distribution. Z score is also used to compare the two different distributions.
**Before and After Z-Score Plot is identical excluding the x-values.**

#### Need for One-hot encoding

<u>One-hot encoding-</u> One hot encoding is a technique that we use to represent categorical variables as numerical values in a machine learning model.

**Advantage:**

1. It allows the use of categorical variables in models that require numerical input.
2. It can improve model performance by providing more information to the model about the categorical variable.
3. It can help to avoid the problem of ordinality, which can occur when a categorical variable has a natural ordering.
